# heatmaps

```python
a=sns.load_dataset('flights')
a
```

```python
x= a.pivot_table(index='year',columns='month',values='passengers')
x
```

```python
x= a.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=True)
```

# divide into small boxes

```python
sns.heatmap(x,cbar=False, linewidth=0.5)
```

```python
sns.heatmap(x,cbar=False, linewidth=0.5,annot=True)
```

# show only integer values

```python
sns.heatmap(x,cbar=False, linewidth=0.5,annot=True,fmt='d')
```

# colors

```python
sns.heatmap(x,cbar=False, linewidth=0.5,annot=True,fmt='d',cmap='viridis') 
```

```python
sns.heatmap(x,cbar=False, linewidth=0.5,annot=True,fmt='d',cmap='summer') 
```

```python
sns.heatmap(x,cbar=False, linewidth=0.5,annot=True,fmt='d',cmap='winter')
```

# cluster maps

```python
sns.clustermap(x)
```

```python
#This parameter specifies the distance metric to be used for hierarchical clustering.
#In this case, 'correlation' is used to compute the correlation distance between the
#rows/columns of the matrix.
sns.clustermap(x, metric='correlation',figsize=(8, 8))
```

```python
sns.clustermap(x, metric='cosine',figsize=(8, 8))
```

```python
sns.clustermap(x, metric='hamming',figsize=(8, 8))
```

```python
sns.clustermap(x, metric='canberra',figsize=(8, 8), )
```

```python
# z_score=0: This parameter standardizes the data in each column (along the row axis)
#by subtracting the mean and dividing by the standard deviation.

sns.clustermap(x, metric='correlation',figsize=(8, 8), z_score=0)
```

```python
#: This parameter displays the data values in each cell of the heatmap.

sns.clustermap(x, z_score=0, annot=True, figsize=(8, 8), metric='correlation')
```

```python
# clustering on the basis of column
sns.clustermap(x, z_score=0, annot=True,row_cluster=False,  figsize=(8, 8),metric='correlation')
```

```python
# clustering on the basis of rows
sns.clustermap(x, z_score=0, annot=True,row_cluster=True,  figsize=(8, 8),metric='correlation')
```

# plotting a displot

```python
sns.displot(data5['age'])
```

```python
sns.displot(data5['age'], kde=True)
```

# histogram

```python

sns.displot(data5['age'], kde=True, element='step', bins=10)
```

```python
sns.displot(data5['age'], kde=False, rug=True)
```

```python

sns.displot(data5['age'], kde=False,element='step',fill=False, stat='density', rug=True)
```
